```python
# Reading the segmented data created from k-Means Clustering
df_read_clustered_data = spark.read.format("csv").option("header", "true").load('/mnt/s3_bucketdata_gr231/clustering_data/pizzacustomer_segmented.csv') 
```

```python
display(df_read_clustered_data)
```

```python
df_read_stream_location_data.createOrReplaceTempView('tab_read_stream_location_data')
df_read_clustered_data.createOrReplaceTempView('tab_read_clustered_data')
```

```python
#Bussiness logic to get if a Customer is eligible for offer
#
spark.sql("""SELECT
          A.CustomerID, 
          A.Gender, 
          A.Age, 
          A.cluster3,
          B.latitude,
          B.longitude,
          ROUND(CAST(B.distance_to_mall_km * 1000 AS FLOAT), 3) AS distance_to_mall_meters,
          CASE WHEN ROUND(CAST(B.distance_to_mall_km * 1000 AS FLOAT), 3) > 0.06
               THEN 'NO OFFER'
               ELSE
               CASE WHEN A.cluster3 = 5
                    THEN "50% OFF OFFER" 
                    WHEN A.cluster3 = 4
                    THEN "40% OFF OFFER" 
                    WHEN A.cluster3 = 3
                    THEN "30% OFF OFFER" 
                    WHEN A.cluster3 = 2
                    THEN "20% OFF OFFER" 
                    WHEN A.cluster3 = 1
                    THEN "10% OFF OFFER" 
                    WHEN A.cluster3 = 0
                    THEN "3% OFF OFFER"
                    END
            END AS offer_status 
          FROM tab_read_clustered_data A
          INNER JOIN tab_read_stream_location_data B
          ON A.CustomerID = B.customer_id
          """).show()
```
