```python
from pyspark.sql.types import StructType, StructField, IntegerType, FloatType, StringType
schema_defined=StructType([StructField('customer_id', IntegerType(), True),
                           StructField('label', IntegerType(), True),
                           StructField('latitude', StringType(), True),
                           StructField('longitude', StringType(), True),
                           StructField('distance_to_mall_km', FloatType(), True)])
```

```python
df_read_stream = spark.readStream.format("csv").schema(schema_defined).option("header", True).option("sep",",").option("cleanSource","delete").load("/mnt/s3_bucketdata_gr231/location_data/")

df_read_stream = df_read_stream.distinct()

display(df_read_stream)
```

```python
df_write_stream = df_read_stream.writeStream.format("parquet").outputMode("append").option("path","/mnt/s3_bucketdata_gr231/stream_write/").option("checkpointLocation","/mnt/s3_bucketdata_gr231/stream_checkpoint/").start().awaitTermination()
```
